```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dt= pd.read_csv('nba.csv')
dt.pivot_table(['Age','Salary'],index=['Position','Team'])
```

```python
a = dt.groupby('Position')
b = lambda g: g['Salary'].max()
c = a.apply(b)
plt.figure()
d=np.arange(len(c))
plt.bar(d,c.values)
plt.xticks(d,c.index)
plt.show()
```

```python
plt.figure(figsize=(15,4))
salary= dt.groupby('Team')
salary_max1 = lambda g: g['Salary'].max() 
salary_max2 = salary.apply(salary_max1)
num1= np.arange(len(salary_max2))
plt.bar(num1,salary_max2.values)
plt.title('Maximum salary for each team')
plt.xticks(num1,salary_max2.index)
plt.show()


plt.figure(figsize=(15,4))
age= dt.groupby('Team')
age_max1 = lambda g: g['Age'].max() 
age_max2 = age.apply(age_max1)
num2= np.arange(len(age_max2))
plt.bar(num2,age_max2.values)
plt.title('Maximum Age for each team')
plt.xticks(num2,age_max2.index)
plt.show()
```

```python
split_name= age_max2.index.str.split(' ',expand=True)
arr=[]
for i in range(len(num2)):
    arr.append(split_name[i][0][0]+split_name[i][1][0]) 
```

```python
plt.figure(figsize=(15,4))
salary= dt.groupby('Team')
salary_max1 = lambda g: g['Salary'].max() 
salary_max2 = salary.apply(salary_max1)
num1= np.arange(len(salary_max2))
plt.bar(num1,salary_max2.values)
plt.title('Maximum salary for each team')
plt.xticks(num1,arr)
plt.show()


plt.figure(figsize=(15,4))
age= dt.groupby('Team')
age_max1 = lambda g: g['Age'].max() 
age_max2 = age.apply(age_max1)
num2= np.arange(len(age_max2))
plt.bar(num2,age_max2.values)
plt.title('Maximum Age for each team')
plt.xticks(num2,arr)
plt.show()
```
