```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
Mpg_data = sns.load_dataset("mpg")
a= Mpg_data['origin']
a.value_counts()
c= Mpg_data['origin'].unique()
c
Mpg_data
```

```python
plt.figure()
a1=np.arange(len(c))
plt.bar(c,a.value_counts())
plt.show()
```

```python
q= Mpg_data[['cylinders', 'mpg', 'acceleration', 'horsepower','origin']]
q
```

```python
sns.pairplot(q, hue='origin', size=2)
```

일본의 자동차 cylinders 갯수가 제일 많고 미국이 평균적으로 많다.
그리고 mpg는 미국이 제일 높으며 일본과 유럽은 비슷하다
accelration은 미국이 월등히 높으며
horsepower는 미국차량이 가장 높고 평균적으로도 높은것을 확인할 수 있다.
