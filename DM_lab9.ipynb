```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

dt=pd.read_csv('nba.csv')
a= dt['Team']=='New York Knicks' 
b= dt['Team']=='Philadelphia 76ers'
c= a|b

dt1= dt[c][['Age','Salary','Team','Position']]
dt1
```

```python
plt.figure(figsize=(10,5))
sns.set(style='whitegrid')
sns.boxplot(data=dt1,x = 'Age',y ='Salary',hue='Position')
plt.grid("")
plt.legend(loc='upper left')
plt.show()
```

'연봉'이 가장 많은 '나이'는 32세 이며 어떠한 포지션이 많이 받는지는 그림을 보고는 알아볼 수가 없다.

```python
sns.set(style='whitegrid')
sns.jointplot(data=dt1, x ='Age',y='Salary',kind='reg')
plt.grid('')
```

선수들의 많은 분포는 20세 에서 30세 사이에 많이 분포 하고 있으며 
나이가 많아 질 수록 평균 연봉이 조금씩 상승하고 있으며 
32세 선수 한명이 독보적으로 연봉이 높은 것을 볼 수 있다.

```python

sns.catplot(data=dt1,x='Age', kind='count',hue='Position',aspect=2.5)
plt.grid('')
```

나이 별로 포지션이 골고루 분포 되어 있지만
PF포지션:20세,22세 에서 2명씩 분포되어 있고 
PG포지션:24세 2명씩
SF포지션:25세 2명씩 분포되어 있는것을 볼 수 있다.
